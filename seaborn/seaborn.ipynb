```python
import seaborn as sns
%matplotlib inline
```

## Distribution Plots

```python
# lets use the tips dataframe built into seaborn
# importing the tips dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.describe()
```

## displot
This allows us to show the distribution of a univariant set of observations
ie. this enables us to do a distribution of a given column

```python
sns.displot(tips['total_bill'],kde=True,bins=30)
```

## jointplot
this allows us to match up two displot for bivariant data. ie two columns

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

the above shows that as you go higher in total bill you go higher in tip

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind="hex")
```

the kind parameter enables us to specify the kind of plot to make. for example kind='hex'
